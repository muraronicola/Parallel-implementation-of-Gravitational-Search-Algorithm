```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


results = pd.read_csv("./data/raw_data_4.csv")
```

```python
#Elaborate data

#Group by comm_sz and pop_size
results_grouped = results.groupby(['comm_sz', 'pop_size']).agg(
    mean_time=('time', 'mean'),
    mean_best_value=('best_value', 'mean'),
).reset_index()


#Find, for each pop_size, calculate the speedup from the comm_sz = 1
for i in [50, 100, 200, 400, 800, 1600, 3200]:
    base_time = results_grouped.loc[(results_grouped['pop_size'] == i) & (results_grouped['comm_sz'] == 1), 'mean_time'].values[0]
    results_grouped.loc[results_grouped['pop_size'] == i, 'speedup'] = base_time / results_grouped.loc[results_grouped['pop_size'] == i, 'mean_time']
    results_grouped.loc[results_grouped['pop_size'] == i, 'efficiency'] = results_grouped.loc[results_grouped['pop_size'] == i, 'speedup'] / results_grouped.loc[results_grouped['pop_size'] == i, 'comm_sz']

print(results_grouped)

#Rename columns
results_grouped.rename(columns={
    'comm_sz': 'n_cores',
    'pop_size': 'n_agents',
    'mean_time': 'time',
    'mean_best_value': 'best_value',
}, inplace=True)

print(results_grouped)

results_grouped.to_csv("./data/avg_data.csv", index=False)
```
