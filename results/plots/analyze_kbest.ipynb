```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import copy

data = pd.read_csv("./data/avg_data.csv")

def get_kbest(pop_size, n_iter_double, t_double):
    return pop_size * ((n_iter_double - t_double) / n_iter_double);
```

```python
pop_size_array = [50, 100, 200, 400, 800, 1600, 3200]
n_iter = 1000


results = pd.DataFrame(columns=["pop_size", "k_best", "iter"])

for pop_size in pop_size_array:
    for i in range(n_iter):
        this_k_best = get_kbest(pop_size, n_iter, i)
        new_result = pd.DataFrame({"pop_size": [pop_size], "k_best": [this_k_best], "iter": [i]})
        results = pd.concat([results, new_result], ignore_index=True)
        
```

```python
figure_1, ax_1 = plt.subplots()  

sns.lineplot(x='iter', y='k_best', data=results, hue="pop_size", ax=ax_1, dashes=False)
```
